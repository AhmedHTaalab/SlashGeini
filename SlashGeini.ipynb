```python
# %%writefile app.py
# import streamlit as st
# import pandas as pd
# import google.generativeai as genai
# import os
# import json
# import re

# os.environ["API_KEY"] = "AIzaSyCYoyXHPR6s7mtIYkz5aJYfaa5mPLrZflA"

# genai.configure(api_key=os.environ["API_KEY"])

# model = genai.GenerativeModel("gemini-1.5-flash")

# with open("./scheme.json", "r") as f:
#     gemini_flash_schema = json.load(f)

# def preprocess_text(text):
#     text = text.lower()
#     text = re.sub(r"\d+", "", text)
#     text = re.sub(r"[^\w\s]", "", text)
#     return " ".join(text)

# def generate_review_grade(review_text):
#     try:
#         schema_str = json.dumps(gemini_flash_schema)
#         prompt = (f"Analyze the following review and category it if it is positive or negative then rate it from 1 to 5 based of the keywords,"
#                   f"if there are keywords: (Good, Not Bad, but) rate it by 3"
#                   f"if there are keywords: (Amazing, Excellent, lovely) rate it by 5"
#                   f"{review_text}")
#         response = model.generate_content(prompt)
#         st.write(f"API Response: {response.text}")  # Debugging statement
#         grade_match = re.search(r"\d(\.\d+)?", response.text)
#         if grade_match:
#             grade = float(grade_match.group())
#             return grade
#         else:
#             st.write(f"No grade found in response: {response.text}")  # Debugging statement
#             return None
#     except Exception as e:
#         st.error(f"Error generating grade for review: {e}")
#         return None

# def calculate_mean_grades():
#     encodings = ['latin1', 'ISO-8859-1', 'cp1252']
#     try:
#         for enc in encodings:
#             try:
#                 df = pd.read_csv("./Test_Data.csv", encoding=enc)
#                 st.write(f"Data columns: {df.columns}")  # Debugging statement
#                 st.write(f"Sample data:\n{df.head()}")  # Debugging statement

#                 # Calculate global average rating
#                 global_avg_rating = df['product_rating'].mean()
#                 min_raters = 35
#                 products = df['Product_Name'].unique()

#                 result = {}
#                 for product in products:
#                     filtered_reviews = df[df["Product_Name"] == product][["product_rating", "product_number_of_rating", "Product_Reviews"]]
#                     if not filtered_reviews.empty:
#                         grades = []
#                         total_weighted_rating = 0
#                         total_mean_grade = 0
#                         st.write(f"Processing product: {product}")  # Debugging statement
#                         for i, row in filtered_reviews.iterrows():
#                             review_text = preprocess_text(row["Product_Reviews"])
#                             st.write(f"Processing review: {review_text}")  # Debugging statement
#                             grade = generate_review_grade(review_text)
#                             if grade:
#                                 grades.append(grade)
#                                 st.write(f"Review {i + 1}: {row['Product_Reviews']}\nGrade: {grade}")
#                                 # Calculate weighted rating
#                                 weighted_rating = ((row['product_rating'] * row['product_number_of_rating']) +
#                                                    (global_avg_rating * min_raters)) / \
#                                                    (row['product_number_of_rating'] + min_raters)
#                                 total_weighted_rating += weighted_rating
#                                 total_mean_grade += grade
#                             else:
#                                 st.write(f"No grade assigned to review: {row['Product_Reviews']}")  # Debugging statement

#                         if grades:
#                             mean_grade = sum(grades) / len(grades)
#                             weighted_avg = total_weighted_rating / len(filtered_reviews)
#                             final_rate = (mean_grade + weighted_avg) / 2
#                             result[product] = {
#                                 "mean_grade": mean_grade,
#                                 "weighted_avg": weighted_avg,
#                                 "final_rate": final_rate,
#                                 "grades": grades
#                             }
#                         else:
#                             result[product] = {
#                                 "mean_grade": None,
#                                 "weighted_avg": None,
#                                 "final_rate": None,
#                                 "grades": []
#                             }
#                     else:
#                         result[product] = {
#                             "mean_grade": None,
#                             "weighted_avg": None,
#                             "final_rate": None,
#                             "grades": []
#                         }

#                 # Convert result to JSON and display
#                 st.write("Grades and Mean Grades for Products:")
#                 st.json(result)
#                 return result

#             except UnicodeDecodeError as e:
#                 st.error(f"Encoding error with {enc}: {e}")
#             except Exception as e:
#                 st.error(f"Error calculating mean grades: {e}")
#                 return None
#     except Exception as e:
#         st.error(f"Error calculating mean grades: {e}")
#         return None

# st.title("Product Review Grader")

# # Calculate and display mean grades
# calculate_mean_grades()


%%writefile app.py
import streamlit as st
import pandas as pd
import google.generativeai as genai
import os
import json
import re
from transformers import pipeline  # For sentiment analysis

os.environ["API_KEY"] = "AIzaSyCYoyXHPR6s7mtIYkz5aJYfaa5mPLrZflA"

genai.configure(api_key=os.environ["API_KEY"])

model = genai.GenerativeModel("gemini-1.5-flash")

with open("./scheme.json", "r") as f:
    gemini_flash_schema = json.load(f)

# Load the sentiment analysis pipeline
sentiment_analyzer = pipeline("sentiment-analysis")

def preprocess_text(text):
    # Convert text to lowercase
    text = text.lower()
    # Remove numbers
    text = re.sub(r"\d+", "", text)
    # Remove punctuation and special characters
    text = re.sub(r"[^\w\s]", "", text)
    # Remove extra spaces
    text = re.sub(r"\s+", " ", text).strip()
    # Remove common stopwords manually
    stopwords = {"the", "is", "in", "at", "on", "a", "an", "and", "or", "for", "to", "of", "with", "that", "by", "it"}
    words = text.split()
    filtered_words = [word for word in words if word not in stopwords]
    # Join the words back into a single string
    return " ".join(filtered_words)

# Function to analyze sentiment
def analyze_sentiment(review_text):
    analysis = sentiment_analyzer(review_text)
    sentiment_label = analysis[0]['label']
    sentiment_score = analysis[0]['score']
    return sentiment_label, sentiment_score

# Modified function to generate review grade
def generate_review_grade(review_text, sentiment_label, sentiment_score):
    try:
        schema_str = json.dumps(gemini_flash_schema)
        prompt = (f"Analyze the following review and categorize it as positive, negative, or neutral, then rate it from 1 to 5 based on the keywords, you MUST grade each review. "
                  f"{review_text}")
        response = model.generate_content(prompt)
        st.write(f"API Response: {response.text}")  # Debugging statement

        grade_match = re.search(r"\d(\.\d+)?", response.text)
        if grade_match:
            grade = float(grade_match.group())

            # Adjust the grade based on sentiment analysis
            if sentiment_label == 'POSITIVE':
                grade = min(grade + 0.5, 5)  # Slightly increase grade for positive sentiment
            elif sentiment_label == 'NEGATIVE':
                grade = max(grade - 0.5, 1)  # Slightly decrease grade for negative sentiment

            return grade
        else:
            st.write(f"No grade found in response: {response.text}")  # Debugging statement
            return None
    except Exception as e:
        st.error(f"Error generating grade for review: {e}")
        return None

def calculate_mean_grades():
    encodings = ['latin1', 'ISO-8859-1', 'cp1252']
    try:
        for enc in encodings:
            try:
                df = pd.read_csv("./English_merged_outputv4.csv", encoding=enc)
                st.write(f"Data columns: {df.columns}")  # Debugging statement
                st.write(f"Sample data:\n{df.head()}")  # Debugging statement

                # Calculate global average rating
                global_avg_rating = df['product_rating'].mean()
                min_raters = 35
                products = df['Product_Name'].unique()

                result = {}
                for product in products:
                    filtered_reviews = df[df["Product_Name"] == product][["product_rating", "product_number_of_rating", "Product_Reviews"]]
                    if not filtered_reviews.empty:
                        grades = []
                        total_weighted_rating = 0
                        total_mean_grade = 0
                        st.write(f"Processing product: {product}")  # Debugging statement
                        for i, row in filtered_reviews.iterrows():
                            review_text = preprocess_text(row["Product_Reviews"])
                            st.write(f"Processing review: {review_text}")  # Debugging statement

                            # Sentiment Analysis
                            sentiment_label, sentiment_score = analyze_sentiment(review_text)
                            st.write(f"Sentiment: {sentiment_label}, Confidence: {sentiment_score}")

                            # Pass the sentiment into the grade generator
                            grade = generate_review_grade(review_text, sentiment_label, sentiment_score)
                            if grade:
                                grades.append(grade)
                                st.write(f"Review {i + 1}: {row['Product_Reviews']}\nGrade: {grade}")
                                # Calculate weighted rating
                                weighted_rating = ((row['product_rating'] * row['product_number_of_rating']) +
                                                   (global_avg_rating * min_raters)) / \
                                                   (row['product_number_of_rating'] + min_raters)
                                total_weighted_rating += weighted_rating
                                total_mean_grade += grade
                            else:
                                st.write(f"No grade assigned to review: {row['Product_Reviews']}")  # Debugging statement

                        if grades:
                            mean_grade = sum(grades) / len(grades)
                            weighted_avg = total_weighted_rating / len(filtered_reviews)
                            final_rate = (mean_grade + weighted_avg) / 2
                            result[product] = {
                                "mean_grade": mean_grade,
                                "weighted_avg": weighted_avg,
                                "final_rate": final_rate,
                                "grades": grades
                            }
                        else:
                            result[product] = {
                                "mean_grade": None,
                                "weighted_avg": None,
                                "final_rate": None,
                                "grades": []
                            }
                    else:
                        result[product] = {
                            "mean_grade": None,
                            "weighted_avg": None,
                            "final_rate": None,
                            "grades": []
                        }

                # Convert result to JSON and display
                st.write("Grades and Mean Grades for Products:")
                st.json(result)
                return result

            except UnicodeDecodeError as e:
                st.error(f"Encoding error with {enc}: {e}")
            except Exception as e:
                st.error(f"Error calculating mean grades: {e}")
                return None
    except Exception as e:
        st.error(f"Error calculating mean grades: {e}")
        return None

st.title("Product Review Grader")

# Calculate and display mean grades
calculate_mean_grades()



```
